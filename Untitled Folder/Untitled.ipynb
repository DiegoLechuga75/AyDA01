```python
def funcionEscalon(k,pesos,b):
    n = -b
    for i in range(len(k)):
        n = n + (k[i] * pesos[i])
    if (n >= 0):
        return 1
    else:
        return 0
```

```python
def entrenarPerceptron(datosEntrada, pesos, b, lamda):
    epocas = 0
    errores = True
    while errores:
        epocas += 1
        print("Epoca: ", epocas)
        errores = False
        for k, y in datosEntrada.items():
            n = funcionEscalon(k, pesos, b)
            if n != y:
                print("Hay ajuste")
                print("p1: ",k[0], " p2: ", k[1])
                errores = True
                e = y - n
                deltaB = -(lamda * e)
                print("b: ", b, " e: ", e, " deltaB: ", deltaB)
                for i in range(len(k)):
                    delta_w = lamda * e * k[i]
                    pesos[i] = pesos[i] + delta_w
                    print("Pesos: ", i, " ", pesos[i])
            else:
                print("No hay ajuste")
                print("p1: ",k[0], " p2: ", k[1])
                e = y - n
                deltaB = -(lamda * e)
                print("b: ", b, " e: ", e, " deltaB: ", deltaB)
        print()
    return pesos, b

def clasificar(entrada, pesos, b):
    return funcionEscalon(entrada, pesos, b)
```

```python
if __name__ == "__main__":
    datosEntrada = {(-7,19):1,(-11,18):1,(-11,5):1,(-7,7):1, (-18,2):1, (-7,0):1, (-8, 17):1, (-20, 9):1, (0, 7):1, (-19,2):1, (17,-4):0, (6,-11):0, (9,-17):0, (5,-11):0, (2,-19):0, (17,-2):0, (8,-15):0, (4,-14):0, (12,-17):0}
    pesos = [.2,.6]
    b = .4
    lamda = .6
    pesos, b = entrenarPerceptron(datosEntrada, pesos, b, lamda)
    print("Pesos finales: ", pesos)
    print("Bias Final: ", b)
    print("Probando perceptron(1,0): "+str(clasificar((1,0),pesos,b)))
```
