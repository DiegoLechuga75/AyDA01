```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("m1.csv", sep=";")
    
x = datos.N
y = datos.Tiempo
    
plt.plot(x, y)
plt.ylabel("Tiempo")
plt.xlabel("N")
plt.title("Burbuja")
plt.legend(('burbuja'),prop={'size':10})
plt.grid()
plt.show()
```

```python
datos = pd.read_csv("m1.csv", sep=";")
    
x = datos.N
y = datos.Tiempo
    
plt.plot(x, y)
plt.ylabel("Tiempo")
plt.xlabel("N")
plt.title("Quick Sort")
plt.legend(('quickSort'),prop={'size':10})
plt.grid()
plt.show()
```

```python
datos = pd.read_csv("m1.csv", sep=";")
datosDos = pd.read_csv("m2.csv", sep=";")
datosTres = pd.read_csv("m3.csv", sep=";")
datosCuatro = pd.read_csv("m4.csv", sep=";")
    
x = datos.N
y = datos.Tiempo
yy = datosDos.Tiempo
yyy = datosTres.Tiempo
yyyy = datosCuatro.Tiempo
    
plt.plot(x, y, x, yy, x, yyy, x, yyyy)
plt.ylabel("Tiempo")
plt.xlabel("N")
plt.title("Burbuja vs Seleccin vs Insercin vs Quick Sort")
plt.legend(('burbuja', 'seleccion', 'insercion', 'quickSort'),prop={'size':10})
plt.grid()
plt.show()
```
